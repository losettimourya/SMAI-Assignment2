```python
# class GMM:
#     def __init__(self, n_components, iterations=100, threshold=0.004, regularization=5e-9):
#         self.n_components = n_components
#         self.means = None
#         self.covariances = None
#         self.prior_probabilities = None
#         self.responsibilities = None
#         self.iterations = iterations
#         self.threshold = threshold
#         self.converged = False
#         self.log_likelihoods = []
#         self.regularization = regularization

#     def initialize_parameters(self, data):
#         n_samples, n_features = data.shape
#         np.random.seed(42)
#         self.responsibilities = np.zeros((n_samples, self.n_components))
#         chosen = np.random.choice(n_samples, self.n_components, replace=False)
#         self.means = data[chosen]
#         self.prior_probabilities = np.full(self.n_components, 1 / self.n_components)
#         self.covariances = np.tile(np.identity(n_features), (self.n_components, 1, 1))
#         return self.means, self.covariances, self.prior_probabilities

#     def fit(self, data):
#         means, covariances, prior_probabilities = self.initialize_parameters(data)
#         log_likelihood = 0
#         for _ in range(self.iterations):
#             new_log_likelihood = self.e_step(data)
#             self.m_step(data)
#             if abs(new_log_likelihood - log_likelihood) <= self.threshold:
#                 self.converged = True
#                 break
#             log_likelihood = new_log_likelihood
#             self.log_likelihoods.append(log_likelihood)
#         return means, covariances, prior_probabilities

#     def get_parameters(self):
#         return self.means, self.covariances, self.prior_probabilities

#     def e_step(self, X):
#         for component in range(self.n_components):
#             self.responsibilities[:, component] = self.prior_probabilities[component] * multivariate_normal(self.means[component], self.covariances[component]).pdf(X)
#         epsilon = 1e-16
#         log_likelihood = np.sum(np.log(np.sum(self.responsibilities, axis=1) + epsilon))
#         self.responsibilities = self.responsibilities / (self.responsibilities.sum(axis=1, keepdims=1) + epsilon)
#         return log_likelihood

#     def m_step(self, X):
#         responsibility_weight_sum = self.responsibilities.sum(axis=0)
#         self.prior_probabilities = responsibility_weight_sum / X.shape[0]
#         self.means = np.dot(self.responsibilities.T, X) / responsibility_weight_sum.reshape(-1, 1)
#         for component in range(self.n_components):
#             self.covariances[component] = np.dot(self.responsibilities[:, component] * (X - self.means[component]).T, (X - self.means[component])) / responsibility_weight_sum[component]
#             self.covariances[component] += self.regularization * np.identity(X.shape[1])


class GMM:
    def __init__(self, n_components, iterations=100, threshold=1e-4, regularization=1e-6):
        self.n_components = n_components
        self.means = None
        self.responsibilities = None
        self.covariances = None
        self.prior_probabilities = None
        self.threshold = threshold
        self.iterations = iterations
        self.converged = False
        self.log_likelihoods = []
        self.regularization = regularization
        # means, covariances, prior_probabilities = self.initialize_parameters(data)
    def initialize_parameters(self, data):
        n_samples, n_features = data.shape
        # np.random.seed(1)
        chosen = np.random.choice(n_samples, self.n_components, replace=False)
        self.responsibilities = np.zeros((n_samples, self.n_components))
        self.means = data[chosen]
        self.prior_probabilities = np.full(self.n_components, 1 / self.n_components)
        self.covariances = np.tile(np.identity(n_features), (self.n_components, 1, 1))
        return self.means, self.covariances, self.prior_probabilities

    def fit(self, data):
        params = self.initialize_parameters(data)
        prev_log_likelihood = 0

        for iter in range(self.iterations):
            log_likelihood = self.e_step(data)
            self.m_step(data)

            if abs(log_likelihood - prev_log_likelihood) <= self.threshold:
                self.converged = True
                break

            prev_log_likelihood = log_likelihood
            self.log_likelihoods.append(log_likelihood)

        return params
    def e_step(self, X):
        log_likelihood = 0
        epsilon = 1e-16

        for component in range(self.n_components):
            self.responsibilities[:, component] = self.prior_probabilities[component] * multivariate_normal(self.means[component], self.covariances[component]).pdf(X)

        log_likelihood = np.sum(np.log(np.sum(self.responsibilities, axis=1) + epsilon))
        self.responsibilities /= (self.responsibilities.sum(axis=1, keepdims=1) + epsilon)

        return log_likelihood


    def m_step(self, X):
        responsibility_weight_sum = np.zeros(self.n_components)
        for component in range(self.n_components):
            responsibility_weight_sum[component] = np.sum(self.responsibilities[:, component])
        self.means = np.dot(self.responsibilities.T, X) / responsibility_weight_sum.reshape(-1, 1)
        self.prior_probabilities = responsibility_weight_sum / X.shape[0]
        # self.prior_probabilities = responsibility_weight_sum
        for component in range(self.n_components):
            self.covariances[component] = np.dot(self.responsibilities[:, component] * (X - self.means[component]).T, (X - self.means[component])) / responsibility_weight_sum[component] + self.regularization * np.identity(X.shape[1])
    def predict(self):
        labels = np.argmax(self.responsibilities, axis=1)
        return labels
    def get_parameters(self):
        return self.means, self.covariances, self.prior_probabilities


```

```python
def calculate_aic_bic(data, n_clusters):
    gmm = GMM(n_components=n_clusters)
    gmm.fit(data)
    # log_likelihood = gmm.score(data)
    log_likelihood = gmm.log_likelihoods[-1]
    num_params = n_clusters * ((data.shape[1]+1)*(data.shape[1]+ 2))/2 - 1  # GMM parameters
    aic = 2 * num_params - 2 * log_likelihood
    bic = np.log(data.shape[0]) * num_params - 2 * log_likelihood
    return aic,bic
```

## Observations

1. The BIC score is typically more cautious than the AIC score, resulting in a stronger penalty for model complexity. This is evident as the BIC score intersects with the AIC score at a lower number of components.

2. Generally, both the AIC and BIC criteria serve as effective tools for determining the optimal number of components in a Gaussian Mixture Model (GMM). Nevertheless, it is essential to acknowledge that these criteria rely on certain data assumptions and may not be universally suitable.

In this specific scenario, the AIC and BIC scores exhibit a local minimum at 9 components, with their respective plots showing a declining trend.
